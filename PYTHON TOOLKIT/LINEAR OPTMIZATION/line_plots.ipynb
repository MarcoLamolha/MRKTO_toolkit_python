```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (5, 5)
```

```python
x_vals = np.arange(-10, 10, 0.3)

x_vals
```

```python
y_vals = 2*x_vals
sns.lineplot(x=x_vals, y=y_vals, color='red')

y_vals = 2*x_vals + 10
sns.lineplot(x=x_vals, y=y_vals, color='blue')

y_vals = 5*x_vals + 10
sns.lineplot(x=x_vals, y=y_vals, color='green')

y_vals = -5*x_vals + 10
sns.lineplot(x=x_vals, y=y_vals, color='purple')
```
