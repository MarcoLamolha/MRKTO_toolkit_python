```python
from pulp import *
```

```python
# Passo 1: definir as variaveis

x = LpVariable('x', 0, 1) # quantidade de milho
y = LpVariable('y', 0, 1) # quantidade de farinha de ossos
```

```python
# Passo 2: definir o objeto
# no nosso caso o objetivo consiste em minimizar o custo

# cria o problema de otimizacao
prob = LpProblem('min_custo_racao_aves', LpMinimize) # define que vamos minimizar

# funcao de custo da racao, que vamos minimizar
prob += (65*x + 30*y)
```

```python
# Passo 3: definir as restricoes (condicoes) do problema

prob += (2*x + 3*y >= 2, 'vitamina_a') # vitamina A
prob += (3*x + 2*y >= 2.5, 'vitamina_b') # vitamina B
prob += (x >= 0.5, 'proteina') # proteina
```

```python
# Passo 3.5: definir os limites das variaveis

prob += (x + y == 1, 'produzir_1kg') # queremos produzir 1kg
# x >= 0 e y >= 0 ja foram definidas na criacao das variaveis
```

```python
print(prob)
```

```python
prob.solve()

print('status:', LpStatus[prob.status])
```

```python
for v in prob.variables():
    print(v.name, '=', v.varValue)
```

```python
print('Custo de 1kg de racao =', prob.objective.value())
```
